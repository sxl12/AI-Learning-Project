```python
# 语音转文本
import openai


file_path = '../../data/audio/audio-01.mp3'
file = open(file_path,"rb")

model = "whisper-1"

whisper_segment = openai.audio.transcriptions.create(
    model=model,
    file=file, # 此处是file非file_path
    response_format="json"
)

print(whisper_segment.text)
```

```python
en_segment = openai.audio.transcriptions.create(
    model=model,
    file=file
 
)

print(en_segment)
```

```python
# 翻译为英语而且prompt必须是英文,根据接口文档试了下目前应该是存在问题不会自动翻译
# https://platform.openai.com/docs/api-reference/audio/createTranslation
en_segment = openai.audio.translations.create(
    model=model,
    file=file
  
)

print(en_segment)
```
